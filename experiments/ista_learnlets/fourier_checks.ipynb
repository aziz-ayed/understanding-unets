```python
%matplotlib nbagg
from fastmri_recon.helpers.utils import gen_mask_tf
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from learning_wavelets.keras_utils.fourier import tf_masked_shifted_normed_fft2d, tf_masked_shifted_normed_ifft2d, tf_shifted_normed_fft2d, tf_shifted_normed_ifft2d
```

```python
image = tf.complex(tf.random.normal([1, 128, 128, 1]), tf.random.normal([1, 128, 128, 1]))
```

```python
mask = gen_mask_tf(image[..., 0], 4)
```

```python
kspace = tf_masked_shifted_normed_fft2d(image[..., 0], mask)
```

```python
x = tf_masked_shifted_normed_ifft2d(kspace, mask)
```

```python
residual = tf_masked_shifted_normed_fft2d(x, mask) - kspace
```

```python
plt.figure()
plt.imshow(np.abs(residual[0]))
```
