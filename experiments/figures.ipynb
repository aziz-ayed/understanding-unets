```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="whitegrid", palette="muted", rc={'figure.figsize': (9, 5), 'image.cmap': 'gray'})
```

```python
df_dncnn_bn = pd.read_csv('dncnn_testing_bn.csv')
df_learnlet_n_samples = pd.read_csv('learnlet_n_samples.csv')
df_unet_n_samples = pd.read_csv('unet_n_samples.csv')
df_original = pd.read_csv('../original_metrics.csv')
df_wavelets = pd.read_csv('../wavelets_metrics.csv')
df_bm3d = pd.read_csv('../bm3d_metrics.csv')
```

```python
df_bm3d
```

```python
plt.figure()
sns.lineplot(x='noise_std', y='keras_psnr', data=df_unet_n_samples.loc[df_unet_n_samples['n_samples'].isna()], label='unet')
sns.lineplot(x='noise_std', y='keras_psnr', data=df_learnlet_n_samples.loc[df_learnlet_n_samples['n_samples'].isna()], label='learnlet')
sns.lineplot(x='noise_std', y='keras_psnr', data=df_dncnn_bn.loc[df_dncnn_bn['bn']], label='dncnn')
sns.lineplot(x='noise_std', y='PSNR', data=df_bm3d, label='bm3d')
sns.lineplot(x='noise_std', y='PSNR', data=df_wavelets, label='wavelets')
plt.legend()
```

```python
plt.figure()
selected_noise_std = 30
sns.lineplot(x='n_samples', y='keras_psnr', data=df_unet_n_samples.loc[df_unet_n_samples['noise_std'] == selected_noise_std].fillna(400), label='unet')
sns.lineplot(x='n_samples', y='keras_psnr', data=df_learnlet_n_samples.loc[df_learnlet_n_samples['noise_std'] == selected_noise_std].fillna(400), label='learnlet')
plt.legend()
```
